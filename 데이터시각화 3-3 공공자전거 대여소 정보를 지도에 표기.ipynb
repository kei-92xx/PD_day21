# 문제 : 공공자전거 대여소 정보를 지도에 표시해 보세요.

```python
import folium
import pandas as pd
import googlemaps
import numpy as np
from folium.plugins import MarkerCluster

gmaps_key = "" # 본인의 키 사용
gmaps = googlemaps.Client(key=gmaps_key)

import json
geo_path = '../data/02. skorea_municipalities_geo_simple.json'
geo_str = json.load(open(geo_path, encoding='utf-8'))

bike = bike = pd.read_excel('../data/공공자전거 대여소 정보.xlsx', engine='openpyxl')

bike_1 = bike[['구분','거치대수']]
bike_1 = bike_1.set_index("구분")
```

```python
mc = MarkerCluster()
map = folium.Map(location=[bike['위도'].mean(),bike['경도'].mean()],
                 zoom_start=11)

map.choropleth(geo_data=geo_str,
               data = bike_1['거치대수'],
               columns = [bike_1.index, bike_1['거치대수']],
               fill_color='RdYlGn_r',
               key_on = 'feature.id')

icon_url = 'http://love.seoul.go.kr/Pds/Board/seoul_news_write/Editor/img201610_20.png'

for n in bike.index:
    if bike['대여소명'][n] !='null':
        icon = folium.Icon(color = 'black', icon = 'hands-bound', prefix='fa')
        iframe = "[대여소명]"+ "</br>" + bike['대여소명'][n]+ "<br>" + "[거치대수]" + str(bike['거치대수'][n])
        popup = folium.Popup(iframe, min_width=100, max_width=130)
        mc.add_child(folium.Marker([bike['위도'][n], bike['경도'][n]],
                                   popup = popup,
                                   icon = folium.CustomIcon(icon_url, icon_size=(40, 40)) ))
            
map.add_child(mc)
map.save('../data/BikeSeoul.html')

map
```
